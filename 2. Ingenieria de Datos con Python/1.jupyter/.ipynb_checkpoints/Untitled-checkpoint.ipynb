```python
import requests

response = requests.get('https://www.eltiempo.com/deportes')

print(response.text)
```

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

print('--------------------------------')
print('titulo')
print(soup.title)

print('--------------------------------')
print('descripcion')
print(soup.select('meta[name=description]'))

print('--------------------------------')
print('content')
print(soup.select('meta[name=description]')[0]['content'])
```

```python
print(soup.title)

cursos = soup.select('.nota')
notas = [cursos['itemtype'] for curso in cursos]

print('----------------------------------')
print('Cursos')
print(notas.count)

print('----------------------------------')
print('Lista de Links')
for nota in notas:
    print(nota)

print('----------------------------------')
print('Finalizo')
    
```
